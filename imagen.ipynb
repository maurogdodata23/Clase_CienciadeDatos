```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
loves = plt.imread("loves.jpeg")
lovesArray = np.array(loves)
np.shape(lovesArray)
plt.imshow(lovesArray)
```

```python
print(type(loves))
```

```python
red = lovesArray[:,:,0]
green = lovesArray[:,:,1]   
blue = lovesArray[:,:,2]
```

```python
plt.imshow(green, cmap="Greys")
```

```python
plt.imshow(red, cmap="Greys")
```

```python
plt.imshow(blue, cmap="Greys")
```

```python
plt.imshow(lovesArray[:, :,0].T, cmap="Greys")
```

```python
plt.imshow(green, cmap="Greens")
plt.title("loves")
plt.axis("off")
plt.show()
```

```python
voltear = np.flip(lovesArray)
plt.imshow(voltear)
plt.title("loves volteada")
plt.axis("off")
plt.show()
```

```python
print(voltear)
```

```python
plt.hist(red.ravel(), bins=256, color='red', alpha=0.5 )
plt.hist(green.ravel(), bins=256, color="green", alpha=0.5)
plt.hist(blue.ravel(), bins=256, color="blue", alpha=0.5)
plt.title("histograma de colores RGB")
plt.xlabel("valor del pixel") 
plt.ylabel("frecuencia")
plt.legend(["rojo", "verde", "azul"])
plt.show()
```
